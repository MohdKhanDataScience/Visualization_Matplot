```python
x=np.linspace(0,1,100)
y=x**2
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y,c='blue')
plt.xlabel('month')
plt.ylabel('profit')
plt.title('Sales Ananalysis')
plt.show()
```

```python
plt.plot(x,y,'x')
```

```python
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
plt.plot(t,t,'r--',label='dayly Sales')
plt.plot(t,t**2,'bs',label='Monthly Sales')
plt.plot(t,t**3,'g^',label='Year Sales')
plt.legend()
plt.title('Sales Analysis')
#plt.grid(color='k')
plt.show()
```

```python
x=np.arange(0,np.pi*2,0.1)
y=np.cos(x)
y2=np.sin(x)
plt.plot(x,y,c='m',linewidth=5,label='Cosine-curve')
plt.plot(x,y2,c='g',linewidth=4,label='Sine-curve')
plt.title('Multiple Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
#vertically:

plt.subplot(211)
plt.plot(x,y,c='g')
plt.title('Cosine-Curve')
plt.subplot(212)
plt.plot(x,y2,c='r')
plt.title('Sine-curve')
plt.show()
```

```python
# Horizontly:

plt.subplot(121)
plt.plot(x,y,c='g')
plt.title('Cosine-Curve')
plt.subplot(122)
plt.plot(x,y2,c='r')
plt.title('Sine-curve')
plt.show()
```

```python
labels=['python','c++','java','c']
sizes=[16,56,75,34]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0,1,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
a = np.array([1,2,3,40,6,13,25,43,34,22])
plt.hist(a, bins = [0,10,20,30,45],color='b')
plt.title("histogram")
plt.show()
```
